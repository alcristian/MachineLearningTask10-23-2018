```python
#importing datasets
#df_2001_2004 = pd.read_csv('Chicago_Crimes_2001_to_2004.csv',error_bad_lines=False)
df_crimes1 = pd.read_csv('Chicago_Crimes_2001_to_2004.csv',error_bad_lines=False)
df_crimes2 = pd.read_csv('Chicago_Crimes_2005_to_2007.csv',error_bad_lines=False)
df_crimes3 = pd.read_csv('Chicago_Crimes_2008_to_2011.csv',error_bad_lines=False)
df_crimes4 = pd.read_csv('Chicago_Crimes_2012_to_2017.csv',error_bad_lines=False)

df_crimedata = pd.concat([df_crimes1, df_crimes2, df_crimes3,df_crimes4], ignore_index=False, axis=0)

del df_crimes1
del df_crimes2
del df_crimes3
del df_crimes4

```

```python
#separando os dados por ano, a partir de 2005 

df_crimedata2005 = df_crimedata[df_crimedata["Year"]==2005]
df_crimedata2006 = df_crimedata[df_crimedata["Year"]==2006]
df_crimedata2007 = df_crimedata[df_crimedata["Year"]==2007]
df_crimedata2008 = df_crimedata[df_crimedata["Year"]==2008]
df_crimedata2009 = df_crimedata[df_crimedata["Year"]==2009]
df_crimedata2010 = df_crimedata[df_crimedata["Year"]==2010]
df_crimedata2011 = df_crimedata[df_crimedata["Year"]==2011]
df_crimedata2012 = df_crimedata[df_crimedata["Year"]==2012]
df_crimedata2013 = df_crimedata[df_crimedata["Year"]==2013]
df_crimedata2014 = df_crimedata[df_crimedata["Year"]==2014]
df_crimedata2015 = df_crimedata[df_crimedata["Year"]==2015]
df_crimedata2016 = df_crimedata[df_crimedata["Year"]==2016]
df_crimedata2017 = df_crimedata[df_crimedata["Year"]==2017]

df_crimedatacartheft2016 = df_crimedata2016[df_crimedata2016["Primary Type"]=='MOTOR VEHICLE THEFT']
#todos os roubos de carro disponiveis
df_crimedatacartheft = df_crimedata[df_crimedata["Primary Type"]=='MOTOR VEHICLE THEFT']
```

```python
#Mapa Crimes Chicago by Ward
#vou utilizar os dados de 2016 porque Chicago implementou o sistema de Boundries em 2015 entao pode haver incoerencias. 

#https://www.cityofchicago.org/city/en/depts/doit/dataset/boundaries_-_wards.html

geo_crimemap = 'Boundaries - Wards (2015-).geojson'
#calcular o numero de incidentes por distrito em 2016
df_WardCrimes2016 = pd.DataFrame(df_crimedata2016['Ward'].value_counts().astype(float))


df_WardCrimes2016 = df_WardCrimes2016.reset_index()
df_WardCrimes2016.columns = ['ward', 'Qtd_Crimes']
df_WardCrimes2016.to_json('Ward_Map.json')

Chicago_COORDINATES = (41.895140898, -87.624255632)

chicago_map1 = folium.Map(location=Chicago_COORDINATES, zoom_start=11)
chicago_map1.choropleth(geo_data = geo_crimemap, 
                data = df_WardCrimes2016,
                columns = ['ward', 'Qtd_Crimes'],
                key_on = 'feature.properties.ward',
                fill_color = 'YlOrRd', 
                fill_opacity = 0.7, 
                line_opacity = 0.2,
                threshold_scale=[0, 1000, 5000, 10000, 15000],
                legend_name = 'Numero de Incidentes por ~police ward~ em 2016')

chicago_map1.save('map_chicago_by_Ward.html')

```

```python
#Mapa de roubo de carros Chicago by Ward /MOTOR VEHICLE THEFT/
#vou utilizar os dados de 2016 porque Chicago implementou o sistema de Boundries em 2015 entao pode haver incoerencias. 

#https://www.cityofchicago.org/city/en/depts/doit/dataset/boundaries_-_wards.html

geo_crimemap2 = 'Boundaries - Wards (2015-).geojson'
#calcular o numero de incidentes por distrito em 2016
df_WardCarTheft2016 = pd.DataFrame(df_crimedatacartheft2016['Ward'].value_counts().astype(float))

df_WardCarTheft2016 = df_WardCarTheft2016.reset_index()
df_WardCarTheft2016.columns = ['ward', 'Qtd_Crimes']
df_WardCarTheft2016.to_json('Ward_Map.json')

Chicago_COORDINATES = (41.895140898, -87.624255632)

chicago_map2 = folium.Map(location=Chicago_COORDINATES, zoom_start=11)
chicago_map2.choropleth(geo_data = geo_crimemap2, 
                data = df_WardCarTheft2016,
                columns = ['ward', 'Qtd_Crimes'],
                key_on = 'feature.properties.ward',
                fill_color = 'YlOrRd', 
                fill_opacity = 0.7, 
                line_opacity = 0.2,
                threshold_scale=[0, 100, 200, 400, 600],
                legend_name = 'Numero de Roubos de carro por ~police ward~ em 2016')

chicago_map2.save('map_car_theft_by_Ward.html')
```

```python
#aplicando FbProphet para predizer os dados utilizando toda a base.

dfall = dfall.rename(columns={'Date': 'ds', 'Qtd_Crimes': 'y'})
CarTheftAll_model = Prophet(interval_width=0.95)
CarTheftAll_model.fit(dfall)
```
